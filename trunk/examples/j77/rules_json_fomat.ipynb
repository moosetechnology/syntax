### INTRO

This notebook gives the format for each rule of f77.semc
It was generated with the code below (stored here in case it will be needed again)

````python
"""
import nbformat
from nbformat import v4 as nbf
import re

def parse_f77_semc(file_path):
    # Define a regular expression pattern to capture the entire rule
    pattern = r'(?<!\*)(<(\d+[a-zA-Z]?[^>]*)>\s*=\s*.*?;)'

    # List to store matched rules
    matched_rules = []

    # Open the file and read its content
    with open(file_path, 'r') as file:
        content = file.read()

        # Use regex to find matches
        matches = re.findall(pattern, content)

        # Add matches to the list
        matched_rules.extend(matches)

    return matched_rules


# Custom sorting function to handle numeric part as integers
def custom_sort_key(entry):
    # Extract numeric part using regular expression
    numeric_part = re.search(r'(\d+)', entry[0]).group(1)
    return int(numeric_part)

file_path = 'f77.semc'
rules = parse_f77_semc(file_path)
rules.sort(key=custom_sort_key)

# generating a file with rules

nb = nbf.new_notebook()

# Create code cells to print rules to every odd cell
cells = [
    nbf.new_raw_cell(f"{rule[0]}")
    if i % 2 != 0 else nbf.new_markdown_cell("```json\n\n```")
    for i, rule in enumerate(rules, 1)
]


nb.cells.extend(cells)

with open('rules_notebook.ipynb', 'w') as f:
    nbformat.write(nb, f)

"""
````
