#### Segment

A SEGMENT is a grouping, defined by the programmer, of simple variables and/or arrays.

It mainly allows you to define TYPES of SEGMENTS; it does not create a segment (see SEGINI). At the same time, a POINTER variable is associated with the segment type; This variable is the DEFAULT POINTER, which can be used to simplify writing references to variables contained in a segment of this type.

```fortran
SEGMENT , FONC
 REAL X(N),Y(M),Z(N,M)
ENDSEGMENT
```

A segment of the `FONC` type is then constituted by the grouping of 3 tables `X`, `Y` and `Z`.
